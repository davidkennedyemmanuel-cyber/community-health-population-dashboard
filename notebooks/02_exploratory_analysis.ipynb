# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Load Data

```python
df=pd.read_csv('../data/cleaned/health_population_cleaned.csv',parse_dates=['year'])
```

## Life Expectancy Trends

```python

plt.figure(figsize=(10,6))
for c in df['country'].unique():
    s=df[df['country']==c]
    plt.plot(s['year'],s['life_expectancy'],label=c)
plt.legend();plt.title('Life Expectancy Trends');plt.show()
```

## Spending vs Mortality

```python

plt.figure(figsize=(8,6))
plt.scatter(df['health_expenditure_pct_gdp'],df['mortality_rate'],alpha=0.6)
plt.title('Health Spending vs Mortality');plt.show()
```
