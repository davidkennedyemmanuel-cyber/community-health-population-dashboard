# Data Cleaning & Feature Engineering

```python
import pandas as pd
import numpy as np
```

## Load Raw Data

```python
df=pd.read_csv('../data/raw/health_population_raw.csv')
df.head()
```

## Overview

```python
df.info()
```

## Handle Missing Values

```python
df=df.fillna(method='ffill')
```

## Convert Year

```python
df['year']=pd.to_datetime(df['year'],format='%Y')
```

## Rolling Averages

```python

df['life_expectancy_3yr_avg']=(
 df.groupby('country')['life_expectancy']
 .rolling(3,1).mean()
 .reset_index(level=0,drop=True)
)
df['mortality_rate_3yr_avg']=(
 df.groupby('country')['mortality_rate']
 .rolling(3,1).mean()
 .reset_index(level=0,drop=True)
)
```

## Save Clean Data

```python
df.to_csv('../data/cleaned/health_population_cleaned.csv',index=False)
```
